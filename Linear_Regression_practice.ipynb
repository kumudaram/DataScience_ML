```python
#separate data used for encoding and nonencoding
# find value count and values in each col
ALLOWED_MAX_UNIQUE_SIZE = 25 # limiting the col size lesser or equ to 25 to be used for encoding
ucols_2_encd = []
for col in df.columns:
    ulen=df[col].nunique()
    #print(col, '=', ulen, df[col].unique())  #print the value in each col along with count
    if ulen <= ALLOWED_MAX_UNIQUE_SIZE:
       ucols_2_encd.append(col)
       print(col, df[col].dtypes)


```

```python
#TO INCLUDE ANY SPECIFIC COL IN THE LIST
ucols_2_encd.append('PUBLISHER')
ucols_2_encd
```

```python
#remove last added column
ucols_2_encd.pop()
ucols_2_encd
```

```python
ucols_2_scl = ['CRITICS_POINTS','USER_POINTS']
ucols_2_scl
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#data pipeline based on data segg as category
cat_pipe_encode = Pipeline(steps=[('ohe', OneHotEncoder())])
```

```python
#data pipeline based on data segg as num
num_pipe_encode = Pipeline(steps=[('scale', StandardScaler())])
```

```python
#comp preprocess pipeline
preprocess = ColumnTransformer(transformers=[('num_encode', num_pipe_encode, ucols_2_scl),
                                            ('cat_encode',cat_pipe_encode, ucols_2_encd),])
```

```python
#creating model pipeline
model_pipeline = Pipeline(steps=[('preprocess', preprocess),
 ('lin_reg', LinearRegression())
 ])
```

```python
from os import X_OK
X = df.drop(columns=[ignore_cols, target_col], axis=1)
#defines axis=1 is column, axis=0 row
X
```

```python
y = df[[target_col]]
y
```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y, test_size=0.1, random_state=42)
train_X.shape, val_X.shape, train_y.shape, val_y.shape
```

```python
train_X
```

```python
train_y
```

```python
model_pipeline.fit(train_X, train_y)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
predicted_train_sales = model_pipeline.predict(train_X)
print('Train RMSE', mean_squared_error(train_y, predicted_train_sales, squared=False) )
```

```python
predicted_val_sales = model_pipeline.predict(val_X)
print('Val RMSE', mean_squared_error(val_y, predicted_val_sales, squared=False) )
```

```python
#Report created as list for various models to check on performance using metrics
rep = []
rep.append(['Plain LR', 2.50, 1.43, 2.50-1.43])
rep
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
preprocess_poly1_cat = ColumnTransformer(
 transformers=[('cat_encode_poly', cat_pipe_encode, ucols_2_encd)]
)

preprocess_poly1_num = ColumnTransformer(
 transformers=[('num_encode_poly', num_pipe_encode, ucols_2_scl)]
)
```

```python
for i in range(2,4):
    model_pipeline_poly1 = Pipeline(
     steps=[('preprocessing_poly1', preprocess_poly1_cat),
      ('poly', PolynomialFeatures(degree=i, include_bias=False)),
            ('preprocessing_poly2', StandardScaler(with_mean=False)), #instead of numPreprocess we have used standardScaler applying to all columns
            ('lin_reg', LinearRegression())]
    )

    model_pipeline_poly1.fit(train_X, train_y)

    predicted_train_poly_sales = model_pipeline_poly1.predict(train_X)
    trmse =  mean_squared_error(train_y,predicted_train_poly_sales,squared=False)
    print('Train Poly-d>', i, 'RMSE', mean_squared_error(train_y, predicted_train_poly_sales, squared=False))

    predicted_val_poly_sales = model_pipeline_poly1.predict(val_X)
    vrmse = mean_squared_error(val_y, predicted_val_poly_sales, squared=False)
    print('Val Poly-d>', i, 'RMSE', mean_squared_error(val_y, predicted_val_poly_sales, squared=False))


    print('-' *60)

    rep.append(['Poly-d=>' +str(i), trmse, vrmse, trmse-vrmse])
```

```python
rep
```

```python
report = pd.DataFrame(rep)
report.columns = ['Model', 'Train_RMSE', 'Val_RMSE', 'Diff']
report
```
